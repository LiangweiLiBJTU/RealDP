```python
# 定义默认分类 pipeline
default_pipeline_cla = {
    "Imputation": 'imp_null',
    "Encoding": 'enc_null',
    "Normalization": 'nor_null',
    "Feature Selection": 'fea_null',
    "Duplication": 'dup_null',
    "Outlier Exclusion": 'out_null',
    "Classification": 'cla_null'
}

# 定义默认回归 pipeline
default_pipeline_reg = {
    "Imputation": 'imp_null',
    "Encoding": 'enc_null',
    "Normalization": 'nor_null',
    "Feature Selection": 'fea_null',
    "Duplication": 'dup_null',
    "Outlier Exclusion": 'out_null',
    "Regression": 'reg_null'
}

# 定义每个任务的算法列表
list1 = ['RAND', 'MF', 'MICE', 'KNN', 'EM', 'MEDIAN', 'MEAN', 'DROP']   # Imputation
list2 = ['OE', 'BE', 'FE', 'CBE', 'LE']                                 # Encoding
list3 = ['ZS', 'DS', 'MM']                                              # Normalization
list4 = ['MR', 'WR', 'LC', 'TB']                                        # Feature Selection
list5 = ['ED', 'AD']                                                    # Duplication
list6 = ['ZSB', 'IQR', 'LOF']                                           # Outlier Exclusion
list7 = ['NB', 'LDA', 'RF', 'LR']                                       # Classification
list8 = ['OLS', 'LASSO', 'RF_REG']                                      # Regression        

# 统一分类 pipeline 顺序
def unify_pipeline_cla(raw_pipeline):
    # 创建一个默认分类 pipeline
    pipeline_cla = default_pipeline_cla.copy()
    
    # 遍历原始 pipeline
    for step in raw_pipeline:
        if step in list1:
            pipeline_cla["Imputation"] = step
        elif step in list2:
            pipeline_cla["Encoding"] = step
        elif step in list3:
            pipeline_cla["Normalization"] = step
        elif step in list4:
            pipeline_cla["Feature Selection"] = step
        elif step in list5:
            pipeline_cla["Duplication"] = step
        elif step in list6:
            pipeline_cla["Outlier Exclusion"] = step
        elif step in list7:
            pipeline_cla["Classification"] = step
        else:
            continue
    
    # 返回按顺序的分类 pipeline
    return [
        pipeline_cla["Imputation"],
        pipeline_cla["Encoding"],
        pipeline_cla["Normalization"],
        pipeline_cla["Feature Selection"],
        pipeline_cla["Duplication"],
        pipeline_cla["Outlier Exclusion"],
        pipeline_cla["Classification"]
    ]

# 统一回归 pipeline 顺序
def unify_pipeline_reg(raw_pipeline):
    # 创建一个默认回归 pipeline
    pipeline_reg = default_pipeline_reg.copy()
    
    # 遍历原始 pipeline
    for step in raw_pipeline:
        if step in list1:
            pipeline_reg["Imputation"] = step
        elif step in list2:
            pipeline_reg["Encoding"] = step
        elif step in list3:
            pipeline_reg["Normalization"] = step
        elif step in list4:
            pipeline_reg["Feature Selection"] = step
        elif step in list5:
            pipeline_reg["Duplication"] = step
        elif step in list6:
            pipeline_reg["Outlier Exclusion"] = step
        elif step in list8:
            pipeline_reg["Regression"] = step
        else:
            continue
    
    # 返回按顺序的回归 pipeline
    return [
        pipeline_reg["Imputation"],
        pipeline_reg["Encoding"],
        pipeline_reg["Normalization"],
        pipeline_reg["Feature Selection"],
        pipeline_reg["Duplication"],
        pipeline_reg["Outlier Exclusion"],
        pipeline_reg["Regression"]
    ]
```
