```python
import pandas as pd
import lightgbm as lgb
```

```python
x_clf = pd.read_csv('dataset\Metafeature.csv')
x_reg = pd.read_csv('dataset\MetafeatureREG.csv')
y_clf = pd.read_csv('dataset\label.csv').drop(columns=['Id'])
y_clf['DatasetName'] = y_clf['DatasetName'].astype('str')
y_clf['DatasetName'] = y_clf['DatasetName'].str.replace(' BankFull', 'BankFull')
y_reg = pd.read_csv('dataset\label_reg.csv').drop(columns=['Id'])
y_clf2 = pd.read_csv('dataset\label_clf2.csv')
y_reg2 = pd.read_csv('dataset\label_reg2.csv')

clf_dataset_names = list(y_clf['DatasetName'].unique())
reg_dataset_names = list(y_reg['DatasetName'].unique())
y_clf1 = y_clf.loc[y_clf.groupby('DatasetName')['EvaluationMetric'].idxmax()].sort_index().reset_index(drop=True)
y_reg1 = y_reg.loc[y_reg.groupby('DatasetName')['EvaluationMetric'].idxmin()].sort_index().reset_index(drop=True)
y_clf1.to_csv('dataset\label_clf1.csv', index=False)
y_reg1.to_csv('dataset\label_reg1.csv', index=False)  
```

```python
y_clf2.head(3)
```

```python
x_clf['type'] = 0
x_reg['type'] = 1
clf_df = pd.concat([x_clf, y_clf2], axis=1)
reg_df = pd.concat([x_reg, y_reg2], axis=1)
df = pd.concat([clf_df, reg_df], axis=0)
df
```

```python
# 对Size列处理
# 删除括号和空格
df['Size'] = df['Size'].str.replace('(', '')
df['Size'] = df['Size'].str.replace(')', '')
df['Size'] = df['Size'].str.replace(' ', '')
# 按照逗号或者星号分割
df['Size'] = df['Size'].str.replace(',', '*')
df['Size'] = df['Size'].str.split('*')
df['size0'] = df['Size'].str[0].astype('int')
df['size1'] = df['Size'].str[1].astype('int')
```

```python
df.shape
```

```python
pipline_split = df['Pipeline'].str.split(',')
for i in range(7):
    df['target' + str(i)] = pipline_split.str[i]
```

```python
df.columns
```

```python
df = df.drop(columns=['Pipeline', 'Size', 'EvaluationMetric', 'Time', 'Website'])
df.to_csv('dataset\data.csv', index=False)
```
